# **Implementação**

```python
# Bibliotecas necessárias 

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

**EROSÃO**

```python


img = cv2.imread('/content/drive/MyDrive/PDI/mulher.jpg', 0)

elemento = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))

img_process = cv2.erode(img, elemento, iterations = 2)

cv2_imshow(img)
cv2_imshow(img_process)

cv2.waitKey(0)
cv2.destroyAllWindows()



```

**DILATAÇÃO**

```python
img = cv2.imread('/content/drive/MyDrive/PDI/mulher.jpg', 0)

elemento = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))

img_process = cv2.dilate(img, elemento, iterations = 2)

#cv2_imshow(img)  # Imagem original
cv2_imshow(img_process)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

**FECHAMENTO**

```python
img = cv2.imread('/content/drive/MyDrive/PDI/mulher.jpg', 0)

elemento = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))

img_process = cv2.morphologyEx(img, cv2.MORPH_CLOSE, elemento)

#cv2_imshow(img)  # Imagem original
cv2_imshow(img_process)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

**ABERTURA**

```python
img = cv2.imread('/content/drive/MyDrive/PDI/mulher.jpg', 0)

elemento = cv2.getStructuringElement(cv2.MORPH_RECT, (7,7))

img_process = cv2.morphologyEx(img, cv2.MORPH_OPEN, elemento)

#cv2_imshow(img)  # Imagem original
cv2_imshow(img_process)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
